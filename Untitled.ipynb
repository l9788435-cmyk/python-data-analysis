```python
# spotify_analysis.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# 1. Load Dataset
# -------------------------------
df = pd.read_csv("spotify.csv")

print("=== FIRST 5 ROWS ===")
print(df.head())

print("\n=== DATA INFO ===")
print(df.info())

print("\n=== STATISTICAL SUMMARY ===")
print(df.describe())

# -------------------------------
# 2. Handle Missing Values
# -------------------------------
print("\n=== MISSING VALUES ===")
print(df.isnull().sum())

# Option 1: Remove missing rows
df = df.dropna()

# -------------------------------
# 3. Top 10 Songs by Popularity
# -------------------------------
top_songs = df.sort_values(
    by="track_popularity",
    ascending=False
).head(10)

print("\n=== TOP 10 SONGS ===")
print(top_songs[['track_name', 'track_artist', 'track_popularity']])

# -------------------------------
# 4. Popularity Distribution
# -------------------------------
plt.figure()
plt.hist(df['track_popularity'], bins=20)
plt.title("Song Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Count")
plt.show()

# -------------------------------
# 5. Correlation Heatmap
# -------------------------------
cols = [
    'track_popularity',
    'danceability',
    'energy',
    'valence',
    'acousticness',
    'tempo'
]

plt.figure(figsize=(8, 6))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()

# -------------------------------
# 6. Top 10 Artists by Avg Popularity
# -------------------------------
artist_popularity = (
    df.groupby('track_artist')['track_popularity']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

print("\n=== TOP 10 ARTISTS BY AVERAGE POPULARITY ===")
print(artist_popularity)

artist_popularity.plot(kind='bar')
plt.title("Top 10 Artists by Average Popularity")
plt.ylabel("Average Popularity")
plt.xlabel("Artist")
plt.xticks(rotation=45, ha='right')
plt.show()

# -------------------------------
# 7. Top Song per Artist
# -------------------------------
top_song_per_artist = df.loc[
    df.groupby('track_artist')['track_popularity'].idxmax(),
    ['track_artist', 'track_name', 'track_popularity']
]

print("\n=== TOP SONG PER ARTIST (SAMPLE) ===")
print(top_song_per_artist.head(15))
```
